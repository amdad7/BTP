```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
t_data = np.array(pd.read_csv('t_25-elems_0to3-time_0.01-timestep_2cos175t-forcing_rrm.csv', header = None))
y_data = np.array(pd.read_csv('y_25-elems_0to3-time_0.01-timestep_2cos175t-forcing_rrm.csv', header = None))

# 175/2*pi = 27.85

print(len(t_data))
print(len(y_data))
print(len(y_data[0]))

T = t_data
Y_1 = y_data[-1]
Y_2 = y_data[int(len(y_data)/2)]

plt.figure(figsize=(40, 5))
plt.plot(T, Y_1, linewidth = 0.8, color = 'black')
plt.title('Deflection at end')
plt.xlabel('Time')
plt.show() 

plt.figure(figsize=(40, 5))
plt.plot(T, Y_2, linewidth = 0.8, color = 'black')
plt.title('Deflection at mid point')
plt.xlabel('Time')
plt.show() 
```
